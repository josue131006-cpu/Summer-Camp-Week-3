## Paso 1: La Alquimia de los Datos (De CSV a Base de Datos SQL)
En el mundo real, los datos no siempre están listos en una base de datos relacional. A veces nos entregan un archivo plano como un `.csv`.

Antes de lanzar comandos SQL, necesitamos crear nuestro servidor local dentro de Colab y cargar nuestro archivo `smash_dataset_clean.csv` dentro de una tabla real llamada `combates`.

> **¡Punto de Control!** Asegúrate de haber arrastrado el archivo `smash_dataset_clean.csv` a la carpeta de la izquierda de Colab antes de correr la siguiente celda.

```python
# Importamos las herramientas que usaremos hoy
import pandas as pd
import sqlite3
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Creamos una conexión a un archivo de base de datos virtual en Colab
conexion = sqlite3.connect("torneo_smash.db")

# 2. Leemos nuestro archivo CSV usando Pandas
df_original = pd.read_csv("smash_dataset_clean.csv", encoding="latin1")

# 3. ¡La Magia! Guardamos toda esa información dentro de una tabla SQL llamada 'combates'
df_original.to_sql("combates", conexion, if_exists="replace", index=False)

print("✅ ¡Éxito total! La base de datos relacional 'torneo_smash.db' ha sido creada y la tabla 'combates' está lista.")
```
